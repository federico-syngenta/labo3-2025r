```python
# ============================================================================
# PREDICCIÓN COMPLETA PARA KAGGLE - MÚLTIPLES MODELOS
# ============================================================================
"""
Proyecto: Predicción de Ventas - Competencia Kaggle
Objetivo: Predecir ventas para febrero 2020 (mes +2)
Modelos: Random Forest, XGBoost, LightGBM, CatBoost, Ensemble
"""

# ============================================================================
# 1. IMPORTS Y CONFIGURACIÓN
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Machine Learning
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, TimeSeriesSplit, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error

# Modelos avanzados
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor

# Series de tiempo
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Utilidades
from tqdm import tqdm
import pickle
import os

print("📦 Librerías cargadas exitosamente!")

# ============================================================================
# 2. CONFIGURACIÓN GLOBAL
# ============================================================================

# Configuración
RANDOM_STATE = 42
TARGET_DATE = '2020-02-01'
VALIDATION_MONTHS = 6  # Últimos 6 meses para validación
OUTPUT_DIR = 'kaggle_predictions'

# Crear directorio de salida
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Configuración visual
plt.style.use('default')
sns.set_palette("husl")

# ============================================================================
# 3. FUNCIONES DE CARGA Y PREPARACIÓN
# ============================================================================

def load_and_prepare_data():
    """Carga y prepara todos los datasets"""
    print("🔄 Cargando datasets...")
    
    try:
        # Cargar datasets
        sales = pd.read_csv("../datasets/sell-in.txt", sep="\t", dtype={"periodo": str})
        stocks = pd.read_csv("../datasets/tb_stocks.txt", sep="\t", dtype={"periodo": str}) 
        product_info = pd.read_csv("../datasets/tb_productos.txt", sep="\t")
        products_to_predict = pd.read_csv('../datasets/product_id_apredecir201912.txt')
        
        # Convertir periodos
        sales['periodo'] = pd.to_datetime(sales['periodo'], format='%Y%m')
        stocks['periodo'] = pd.to_datetime(stocks['periodo'], format='%Y%m')
        
        print(f"✅ Sales: {sales.shape[0]:,} filas")
        print(f"✅ Stocks: {stocks.shape[0]:,} filas") 
        print(f"✅ Products: {product_info.shape[0]:,} productos")
        print(f"✅ Products to predict: {len(products_to_predict):,} productos")
        
        return sales, stocks, product_info, products_to_predict
        
    except Exception as e:
        print(f"❌ Error cargando datos: {e}")
        return None, None, None, None

def load_indec_data():
    """Carga y procesa datos del INDEC"""
    print("🇦🇷 Cargando datos del IPC INDEC...")
    
    try:
        INDEC = pd.read_csv('../datasets/serie_ipc_aperturas.csv', sep=';', encoding='latin-1')
        INDEC['periodo'] = INDEC['periodo'].astype(str)
        
        INDEC_filtered = INDEC[
            (INDEC['periodo'] >= '201701') & 
            (INDEC['periodo'] <= '202012') & 
            (INDEC['Descripcion_aperturas'] == 'Nivel general')
        ].copy()
        
        def clean_and_convert(value):
            if isinstance(value, str):
                try:
                    numbers = value.split()
                    cleaned_numbers = [float(num.replace(',', '.')) for num in numbers if num]
                    return np.mean(cleaned_numbers) if cleaned_numbers else np.nan
                except:
                    return np.nan
            return value
        
        INDEC_filtered['v_m_IPC'] = INDEC_filtered['v_m_IPC'].apply(clean_and_convert)
        INDEC_filtered = INDEC_filtered.dropna(subset=['v_m_IPC'])
        INDEC_processed = INDEC_filtered.groupby('periodo')['v_m_IPC'].mean().reset_index()
        INDEC_processed['periodo'] = pd.to_datetime(INDEC_processed['periodo'], format='%Y%m')
        
        print(f"✅ IPC INDEC procesado: {len(INDEC_processed)} períodos")
        return INDEC_processed
        
    except Exception as e:
        print(f"❌ Error procesando INDEC: {e}")
        return None

def create_advanced_features(sales, stocks, product_info, indec_data):
    """Crea features avanzadas combinando todos los datasets"""
    print("🔧 Creando features avanzadas...")
    
    # Merge inicial
    data = sales.copy()
    
    # Agregar información de productos
    if product_info is not None:
        data = data.merge(product_info, on='product_id', how='left')
    
    # Agregar stocks
    if stocks is not None:
        data = data.merge(stocks, on=['periodo', 'product_id'], how='left')
    
    # Agregar IPC
    if indec_data is not None:
        data = data.merge(indec_data, on='periodo', how='left')
    
    # FEATURE ENGINEERING COMPLETO
    print("📊 Aplicando feature engineering...")
    
    # 1. Features temporales
    data['year'] = data['periodo'].dt.year
    data['month'] = data['periodo'].dt.month
    data['quarter'] = data['periodo'].dt.quarter
    data['is_weekend'] = data['periodo'].dt.dayofweek.isin([5, 6]).astype(int)
    data['is_month_start'] = data['periodo'].dt.is_month_start.astype(int)
    data['is_month_end'] = data['periodo'].dt.is_month_end.astype(int)
    
    # 2. Features de estacionalidad
    data['sin_month'] = np.sin(2 * np.pi * data['month'] / 12)
    data['cos_month'] = np.cos(2 * np.pi * data['month'] / 12)
    
    # 3. Lags y rolling windows
    print("🔄 Creando lags y rolling features...")
    lags = [1, 2, 3, 6, 12, 24]
    for lag in tqdm(lags, desc="Creando lags"):
        print(f"   → Procesando lag {lag}...")
        data[f'sales_lag_{lag}'] = data.groupby(['product_id', 'customer_id'])['tn'].shift(lag)

    print("🔄 Creando rolling features...")
    windows = [3, 6, 12]
    operations = ['mean', 'std', 'ewma']
    total_ops = len(windows) * len(operations)

    with tqdm(total=total_ops, desc="Rolling features") as pbar:
        for window in windows:
            print(f"   → Rolling mean {window}...")
            data[f'sales_rolling_mean_{window}'] = data.groupby(['product_id', 'customer_id'])['tn'].transform(
                lambda x: x.rolling(window, min_periods=1).mean()
            )
            pbar.update(1)
            
            print(f"   → Rolling std {window}...")
            data[f'sales_rolling_std_{window}'] = data.groupby(['product_id', 'customer_id'])['tn'].transform(
                lambda x: x.rolling(window, min_periods=1).std()
            )
            pbar.update(1)
            
            print(f"   → EWMA {window}...")
            data[f'sales_ewma_{window}'] = data.groupby(['product_id', 'customer_id'])['tn'].transform(
                lambda x: x.ewm(span=window, min_periods=1).mean()
            )
            pbar.update(1)


    # 4. Agregaciones por categorías
    print("🏷️ Creando agregaciones...")
    categorical_cols = ['product_id', 'customer_id']
    if 'brand' in data.columns:
        categorical_cols.extend(['brand', 'cat1', 'cat2', 'cat3'])
    
    for cat in tqdm(categorical_cols, desc="Agregaciones por categoría"):
        if cat in data.columns:
            data[f'{cat}_mean'] = data.groupby(cat)['tn'].transform('mean')
            data[f'{cat}_std'] = data.groupby(cat)['tn'].transform('std')
            data[f'{cat}_median'] = data.groupby(cat)['tn'].transform('median')
    
    # 5. Features de stock (si disponible)
    print("📦 Creando features de stock...")  
    if 'stock_final' in data.columns:
        data['stock_turnover'] = data['tn'] / (data['stock_final'] + 1e-6)
        data['days_of_stock'] = data['stock_final'] / (data['tn'] + 1e-6) * 30
    
    # 6. Features de IPC (si disponible)
    print("💰 Creando features de IPC...")  # Agregar esta línea
    if 'v_m_IPC' in data.columns:
        for lag in [1, 2, 3]:
            data[f'ipc_lag_{lag}'] = data['v_m_IPC'].shift(lag)
        data['ipc_rolling_3'] = data['v_m_IPC'].rolling(3, min_periods=1).mean()
        data['sales_ipc_interaction'] = data['tn'] * data['v_m_IPC']
    
    # 7. Features de tendencia
    data['sales_diff'] = data.groupby(['product_id', 'customer_id'])['tn'].diff()
    data['sales_pct_change'] = data.groupby(['product_id', 'customer_id'])['tn'].pct_change()
    data['yoy_growth'] = data.groupby(['product_id', 'customer_id'])['tn'].pct_change(periods=12)
    
    # 8. Acumulados
    data['cumulative_sales'] = data.groupby(['product_id', 'customer_id'])['tn'].cumsum()
    
    # Limpieza final
    data = data.replace([np.inf, -np.inf], np.nan)
    data = data.fillna(0)
    
    print(f"✅ Features creadas. Shape final: {data.shape}")
    return data

# ============================================================================
# 4. AGREGACIÓN Y PREPARACIÓN PARA MODELOS
# ============================================================================

def aggregate_for_modeling(data, products_to_predict):
    """Agrega datos por producto y período para modelado"""
    print("📊 Agregando datos por producto...")
    
    # Definir features relevantes
    relevant_features = [
        'product_id', 'periodo', 'tn',
        # Temporales
        'year', 'month', 'quarter', 'is_weekend', 'is_month_start', 'is_month_end',
        'sin_month', 'cos_month',
        # Lags
        'sales_lag_1', 'sales_lag_2', 'sales_lag_3', 'sales_lag_6', 'sales_lag_12',
        # Rolling
        'sales_rolling_mean_3', 'sales_rolling_mean_6', 'sales_rolling_mean_12',
        'sales_rolling_std_3', 'sales_rolling_std_6', 'sales_rolling_std_12',
        'sales_ewma_3', 'sales_ewma_6', 'sales_ewma_12',
        # Agregaciones
        'product_id_mean', 'product_id_std', 'product_id_median',
        'customer_id_mean', 'customer_id_std', 'customer_id_median',
        # Diferencias
        'sales_diff', 'sales_pct_change', 'yoy_growth',
        # Acumulados
        'cumulative_sales',
        # Categóricas
        'cat1', 'cat2', 'cat3', 'brand', 'sku_size',
        # Stock (si existe)
        'stock_final', 'stock_turnover', 'days_of_stock',
        # IPC (si existe)
        'v_m_IPC', 'ipc_lag_1', 'ipc_lag_2', 'ipc_lag_3', 
        'ipc_rolling_3', 'sales_ipc_interaction'
    ]
    
    # Filtrar solo columnas que existen
    existing_features = [col for col in relevant_features if col in data.columns]
    data_filtered = data[existing_features]
    
    # Definir agregaciones
    agg_dict = {}
    for col in data_filtered.columns:
        if col in ['product_id', 'periodo']:
            continue
        elif col == 'tn':
            agg_dict[col] = 'sum'
        elif col in ['cat1', 'cat2', 'cat3', 'brand', 'sku_size']:
            agg_dict[col] = 'first'
        else:
            agg_dict[col] = 'mean'
    
    # Agregar
    data_grouped = data_filtered.groupby(['product_id', 'periodo']).agg(agg_dict).reset_index()
    
    # Filtrar solo productos a predecir
    data_grouped = data_grouped[data_grouped['product_id'].isin(products_to_predict['product_id'])]
    data_grouped = data_grouped.sort_values(['product_id', 'periodo'])
    
    print(f"✅ Datos agregados: {data_grouped.shape}")
    return data_grouped

# ============================================================================
# 5. VALIDACIÓN TEMPORAL
# ============================================================================

def create_temporal_splits(data, validation_months=3):
    """Crea splits temporales para validación"""
    print(f"📅 Creando splits temporales ({validation_months} meses de validación)...")
    
    # Encontrar fecha de corte
    max_date = data['periodo'].max()
    cutoff_date = max_date - pd.DateOffset(months=validation_months)
    
    # Crear splits
    train_data = data[data['periodo'] <= cutoff_date].copy()
    val_data = data[data['periodo'] > cutoff_date].copy()
    
    print(f"📊 Training: {len(train_data):,} registros (hasta {cutoff_date.strftime('%Y-%m')})")
    print(f"📊 Validation: {len(val_data):,} registros (desde {(cutoff_date + pd.DateOffset(months=1)).strftime('%Y-%m')})")
    
    return train_data, val_data

# ============================================================================
# 6. MODELOS DE MACHINE LEARNING
# ============================================================================

class ModelEvaluator:
    """Clase para evaluar múltiples modelos"""
    
    def __init__(self, random_state=42):
        self.random_state = random_state
        self.models = {}
        self.results = {}
        self.feature_importances = {}
    
    def prepare_features(self, data):
        """Prepara features y target"""
        # Separar categóricas y numéricas
        categorical_features = ['cat1', 'cat2', 'cat3', 'brand', 'sku_size']
        categorical_features = [col for col in categorical_features if col in data.columns]
        
        numerical_features = [col for col in data.columns 
                             if col not in ['product_id', 'periodo', 'tn'] + categorical_features]
        
        features = data[categorical_features + numerical_features]
        target = data['tn']
        
        return features, target, categorical_features, numerical_features
    
    def create_preprocessor(self, categorical_features, numerical_features):
        """Crea preprocessor para pipeline"""
        return ColumnTransformer(
            transformers=[
                ('num', Pipeline([
                    ('imputer', SimpleImputer(strategy='mean')),
                    ('scaler', StandardScaler())
                ]), numerical_features),
                ('cat', Pipeline([
                    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
                    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
                ]), categorical_features)
            ],
            remainder='drop'
        )
    
    def initialize_models(self):
        """Inicializa todos los modelos"""
        self.models = {
            'RandomForest': RandomForestRegressor(
                n_estimators=500,
                max_depth=15,
                min_samples_split=5,
                min_samples_leaf=2,
                random_state=self.random_state,
                n_jobs=-1
            ),
            'XGBoost': xgb.XGBRegressor(
                n_estimators=500,
                max_depth=8,
                learning_rate=0.1,
                subsample=0.8,
                colsample_bytree=0.8,
                random_state=self.random_state,
                n_jobs=-1
            ),
            'LightGBM': lgb.LGBMRegressor(
                n_estimators=500,
                max_depth=8,
                learning_rate=0.1,
                subsample=0.8,
                colsample_bytree=0.8,
                random_state=self.random_state,
                n_jobs=-1,
                verbose=-1
            ),
            'CatBoost': CatBoostRegressor(
                iterations=500,
                depth=8,
                learning_rate=0.1,
                random_seed=self.random_state,
                verbose=False
            ),
            'GradientBoosting': GradientBoostingRegressor(
                n_estimators=500,
                max_depth=8,
                learning_rate=0.05,
                subsample=0.9,
                random_state=self.random_state
            )
        }
    
    def evaluate_model(self, model_name, model, X_train, X_val, y_train, y_val, preprocessor):
        """Evalúa un modelo individual"""
        print(f"🔄 Evaluando {model_name}...")
        
        # Crear pipeline
        if model_name == 'CatBoost':
            # CatBoost maneja categóricas nativamente
            pipeline = model
            X_train_processed = preprocessor.fit_transform(X_train)
            X_val_processed = preprocessor.transform(X_val)
        else:
            pipeline = Pipeline([
                ('preprocessor', preprocessor),
                ('regressor', model)
            ])
            X_train_processed = X_train
            X_val_processed = X_val
        
        # Entrenar
        pipeline.fit(X_train_processed, y_train)
        
        # Predecir
        y_pred = pipeline.predict(X_val_processed)
        
        # Calcular métricas
        mae = mean_absolute_error(y_val, y_pred)
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        r2 = r2_score(y_val, y_pred)
        
        # MAPE mejorado con threshold
        mask = y_val > 0.1  # Solo calcular MAPE para ventas > 0.1
        mape = np.mean(np.abs((y_val[mask] - y_pred[mask]) / y_val[mask])) * 100 if mask.sum() > 0 else 0
        
        # Guardar resultados
        self.results[model_name] = {
            'MAE': mae,
            'RMSE': rmse,
            'R2': r2,
            'MAPE': mape,
            'pipeline': pipeline
        }
        
        # Guardar feature importances si disponible
        if hasattr(pipeline.named_steps.get('regressor', model), 'feature_importances_'):
            self.feature_importances[model_name] = pipeline.named_steps.get('regressor', model).feature_importances_
        
        print(f"✅ {model_name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.3f}, MAPE: {mape:.1f}%")
        return pipeline
    
    def evaluate_all_models(self, train_data, val_data):
        """Evalúa todos los modelos"""
        print("🚀 Iniciando evaluación de modelos...")
        
        # Preparar datos
        X_train, y_train, cat_features, num_features = self.prepare_features(train_data)
        X_val, y_val, _, _ = self.prepare_features(val_data)
        
        # Crear preprocessor
        preprocessor = self.create_preprocessor(cat_features, num_features)
        
        # Inicializar modelos
        self.initialize_models()
        
        # Evaluar cada modelo
        for model_name, model in tqdm(self.models.items(), desc="Evaluando modelos"):
            try:
                self.evaluate_model(model_name, model, X_train, X_val, y_train, y_val, preprocessor)
            except Exception as e:
                print(f"❌ Error con {model_name}: {e}")
    
    def get_best_model(self):
        """Obtiene el mejor modelo basado en RMSE"""
        if not self.results:
            return None
        
        best_model_name = min(self.results.keys(), key=lambda x: self.results[x]['RMSE'])
        return best_model_name, self.results[best_model_name]['pipeline']
    
    def print_results_summary(self):
        """Imprime resumen de resultados"""
        print("\n" + "="*60)
        print("📊 RESUMEN DE RESULTADOS")
        print("="*60)
        
        results_df = pd.DataFrame(self.results).T
        results_df = results_df.sort_values('RMSE')
        
        print(results_df[['MAE', 'RMSE', 'R2', 'MAPE']].round(3))
        
        best_model = results_df.index[0]
        print(f"\n🏆 MEJOR MODELO: {best_model}")
        print(f"   RMSE: {results_df.loc[best_model, 'RMSE']:.3f}")
        print(f"   MAE: {results_df.loc[best_model, 'MAE']:.3f}")
        print(f"   R2: {results_df.loc[best_model, 'R2']:.3f}")
        
        return results_df

# ============================================================================
# 7. PREDICCIONES FINALES
# ============================================================================

def make_final_predictions(best_pipeline, data, products_to_predict, target_date='2020-02-01'):
    """Hace predicciones finales para todos los productos"""
    print(f"🎯 Generando predicciones para {target_date}...")
    
    predictions = []
    errors = []
    
    for product_id in tqdm(products_to_predict['product_id'], desc="Prediciendo"):
        try:
            # Filtrar datos del producto
            product_data = data[data['product_id'] == product_id].copy()
            
            if len(product_data) < 3:
                # Fallback inteligente basado en categoría similar
                category = product_data['cat1'].iloc[0] if len(product_data) > 0 else None
                if category and category in data['cat1'].values:
                    similar_avg = data[data['cat1'] == category]['tn'].mean()
                    fallback_value = max(0, similar_avg * 0.1)  # 10% del promedio de la categoría
                else:
                    fallback_value = 0.0
                
                predictions.append({
                    'product_id': product_id,
                    'tn': fallback_value,
                    'method': 'category_fallback'
                })
                continue
            
            # Preparar features para predicción
            last_row = product_data.iloc[-1].copy()
            
            # Actualizar features temporales para febrero 2020
            target_dt = pd.to_datetime(target_date)
            last_row['year'] = target_dt.year
            last_row['month'] = target_dt.month
            last_row['quarter'] = target_dt.quarter
            last_row['sin_month'] = np.sin(2 * np.pi * target_dt.month / 12)
            last_row['cos_month'] = np.cos(2 * np.pi * target_dt.month / 12)
            
            # Crear DataFrame para predicción
            X_pred = pd.DataFrame([last_row])
            
            # Filtrar solo features necesarias
            feature_cols = [col for col in X_pred.columns 
                           if col not in ['product_id', 'periodo', 'tn']]
            X_pred = X_pred[feature_cols]
            
            # Hacer predicción
            pred = best_pipeline.predict(X_pred)[0]
            
            predictions.append({
                'product_id': product_id,
                'tn': max(0, pred),  # No permitir predicciones negativas
                'method': 'model'
            })
            
        except Exception as e:
            errors.append({'product_id': product_id, 'error': str(e)})
            # Fallback
            predictions.append({
                'product_id': product_id,
                'tn': 0.0,
                'method': 'error_fallback'
            })
    
    predictions_df = pd.DataFrame(predictions)
    
    print(f"✅ Predicciones completadas:")
    print(f"   Total: {len(predictions_df)}")
    print(f"   Con modelo: {len(predictions_df[predictions_df['method'] == 'model'])}")
    print(f"   Fallback: {len(predictions_df[predictions_df['method'] != 'model'])}")
    print(f"   Errores: {len(errors)}")
    
    return predictions_df, errors

# ============================================================================
# 8. VISUALIZACIONES
# ============================================================================

def plot_model_comparison(results_df):
    """Visualiza comparación de modelos"""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # MAE
    axes[0,0].bar(results_df.index, results_df['MAE'], color='steelblue', alpha=0.7)
    axes[0,0].set_title('Mean Absolute Error (MAE)')
    axes[0,0].set_ylabel('MAE')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # RMSE
    axes[0,1].bar(results_df.index, results_df['RMSE'], color='darkorange', alpha=0.7)
    axes[0,1].set_title('Root Mean Square Error (RMSE)')
    axes[0,1].set_ylabel('RMSE')
    axes[0,1].tick_params(axis='x', rotation=45)
    
    # R2
    axes[1,0].bar(results_df.index, results_df['R2'], color='green', alpha=0.7)
    axes[1,0].set_title('R² Score')
    axes[1,0].set_ylabel('R²')
    axes[1,0].tick_params(axis='x', rotation=45)
    
    # MAPE
    axes[1,1].bar(results_df.index, results_df['MAPE'], color='red', alpha=0.7)
    axes[1,1].set_title('Mean Absolute Percentage Error (MAPE)')
    axes[1,1].set_ylabel('MAPE (%)')
    axes[1,1].tick_params(axis='x', rotation=45)
    
    plt.suptitle('Comparación de Modelos', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/model_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_predictions_distribution(predictions_df):
    """Visualiza distribución de predicciones"""
    plt.figure(figsize=(12, 6))
    
    # Histograma de predicciones
    plt.subplot(1, 2, 1)
    plt.hist(predictions_df['tn'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title('Distribución de Predicciones')
    plt.xlabel('Toneladas predichas')
    plt.ylabel('Frecuencia')
    plt.grid(True, alpha=0.3)
    
    # Box plot
    plt.subplot(1, 2, 2)
    plt.boxplot(predictions_df['tn'])
    plt.title('Box Plot de Predicciones')
    plt.ylabel('Toneladas predichas')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/predictions_distribution.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Estadísticas
    print("\n📊 ESTADÍSTICAS DE PREDICCIONES:")
    print(f"Total de productos: {len(predictions_df)}")
    print(f"Predicciones > 0: {(predictions_df['tn'] > 0).sum()}")
    print(f"Media: {predictions_df['tn'].mean():.2f}")
    print(f"Mediana: {predictions_df['tn'].median():.2f}")
    print(f"Máximo: {predictions_df['tn'].max():.2f}")
    print(f"Mínimo: {predictions_df['tn'].min():.2f}")

# ============================================================================
# 9. GUARDADO DE RESULTADOS
# ============================================================================

def save_results(predictions_df, results_df, best_model_name, errors):
    """Guarda todos los resultados"""
    print("💾 Guardando resultados...")
    
    # Predicciones para Kaggle (formato requerido)
    kaggle_submission = predictions_df[['product_id', 'tn']].copy()
    kaggle_submission.to_csv(f'{OUTPUT_DIR}/kaggle_submission.csv', index=False)
    
    # Predicciones detalladas
    predictions_df.to_csv(f'{OUTPUT_DIR}/detailed_predictions.csv', index=False)
    
    # Resultados de modelos
    results_df.to_csv(f'{OUTPUT_DIR}/model_results.csv')
    
    # Errores
    if errors:
        pd.DataFrame(errors).to_csv(f'{OUTPUT_DIR}/prediction_errors.csv', index=False)
    
    # Resumen
    summary = {
        'best_model': best_model_name,
        'total_products': len(predictions_df),
        'predictions_with_model': (predictions_df['method'] == 'model').sum(),
        'fallback_predictions': (predictions_df['method'] != 'model').sum(),
        'total_errors': len(errors),
        'mean_prediction': predictions_df['tn'].mean(),
        'median_prediction': predictions_df['tn'].median(),
        'max_prediction': predictions_df['tn'].max(),
        'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    }
    
    with open(f'{OUTPUT_DIR}/summary.json', 'w') as f:
        import json
        json.dump(summary, f, indent=2, default=str)
    
    print(f"✅ Resultados guardados en directorio: {OUTPUT_DIR}")
    print(f"   📄 kaggle_submission.csv: {len(kaggle_submission)} predicciones")
    print(f"   📄 detailed_predictions.csv: Predicciones detalladas")
    print(f"   📄 model_results.csv: Resultados de todos los modelos")
    print(f"   📄 summary.json: Resumen ejecutivo")

# ============================================================================
# 10. FUNCIÓN PRINCIPAL
# ============================================================================

def main():
    """Función principal que ejecuta todo el pipeline"""
    print("🚀 INICIANDO PREDICCIÓN COMPLETA PARA KAGGLE")
    print("="*60)
    
    start_time = datetime.now()
    
    try:
        # 1. Cargar datos
        print("\n1️⃣ CARGA DE DATOS")
        sales, stocks, product_info, products_to_predict = load_and_prepare_data()
        if sales is None:
            print("❌ Error en carga de datos. Abortando...")
            return
        
        # 2. Cargar datos INDEC
        print("\n2️⃣ CARGA DE DATOS INDEC")
        indec_data = load_indec_data()
        
        # 3. Feature Engineering
        print("\n3️⃣ FEATURE ENGINEERING")
        data_with_features = create_advanced_features(sales, stocks, product_info, indec_data)
        
        # 4. Agregación para modelado
        print("\n4️⃣ AGREGACIÓN DE DATOS")
        model_data = aggregate_for_modeling(data_with_features, products_to_predict)
        
        # 5. Splits temporales
        print("\n5️⃣ CREACIÓN DE SPLITS TEMPORALES")
        train_data, val_data = create_temporal_splits(model_data, VALIDATION_MONTHS)
        
        # 6. Evaluación de modelos
        print("\n6️⃣ EVALUACIÓN DE MODELOS")
        evaluator = ModelEvaluator(random_state=RANDOM_STATE)
        evaluator.evaluate_all_models(train_data, val_data)
        
        # 7. Resultados de evaluación
        print("\n7️⃣ ANÁLISIS DE RESULTADOS")
        results_df = evaluator.print_results_summary()
        
        # 8. Mejor modelo
        print("\n8️⃣ SELECCIÓN DEL MEJOR MODELO")
        best_model_name, best_pipeline = evaluator.get_best_model()
        
        # 9. Reentrenamiento con todos los datos
        print("\n9️⃣ REENTRENAMIENTO CON DATOS COMPLETOS")
        print("🔄 Reentrenando el mejor modelo con todos los datos disponibles...")
        from tqdm import tqdm
        # Preparar datos completos para reentrenamiento
        X_full, y_full, cat_features, num_features = evaluator.prepare_features(model_data)
        best_pipeline.fit(X_full, y_full)
        
        # 10. Predicciones finales
        print("\n🔟 PREDICCIONES FINALES")
        predictions_df, errors = make_final_predictions(
            best_pipeline, model_data, products_to_predict, TARGET_DATE
        )
        
        # 11. Visualizaciones
        print("\n1️⃣1️⃣ GENERACIÓN DE VISUALIZACIONES")
        plot_model_comparison(results_df)
        plot_predictions_distribution(predictions_df)
        
        # 12. Guardar resultados
        print("\n1️⃣2️⃣ GUARDADO DE RESULTADOS")
        save_results(predictions_df, results_df, best_model_name, errors)
        
        # Resumen final
        end_time = datetime.now()
        execution_time = end_time - start_time
        
        print("\n" + "="*60)
        print("🎉 PROCESO COMPLETADO EXITOSAMENTE")
        print("="*60)
        print(f"⏱️  Tiempo de ejecución: {execution_time}")
        print(f"🎯 Mejor modelo: {best_model_name}")
        print(f"📊 Total predicciones: {len(predictions_df)}")
        print(f"💾 Archivos guardados en: {OUTPUT_DIR}/")
        print("\n📋 ARCHIVOS GENERADOS:")
        print("   📄 kaggle_submission.csv - Para subir a Kaggle")
        print("   📄 detailed_predictions.csv - Predicciones detalladas")
        print("   📄 model_results.csv - Comparación de modelos")
        print("   📄 summary.json - Resumen ejecutivo")
        print("   📊 model_comparison.png - Gráfico comparativo")
        print("   📊 predictions_distribution.png - Distribución de predicciones")
        
    except Exception as e:
        print(f"\n❌ ERROR CRÍTICO: {e}")
        import traceback
        traceback.print_exc()
        return

# ============================================================================
# 11. FUNCIONES ADICIONALES DE ANÁLISIS
# ============================================================================

def analyze_feature_importance(evaluator, top_n=20):
    """Analiza la importancia de features del mejor modelo"""
    print(f"\n📊 ANÁLISIS DE IMPORTANCIA DE FEATURES (TOP {top_n})")
    print("="*60)
    
    if not evaluator.feature_importances:
        print("❌ No hay información de importancia de features disponible")
        return
    
    # Obtener el mejor modelo
    best_model_name = min(evaluator.results.keys(), 
                         key=lambda x: evaluator.results[x]['RMSE'])
    
    if best_model_name in evaluator.feature_importances:
        importances = evaluator.feature_importances[best_model_name]
        
        # Crear DataFrame con importancias (nota: esto es simplificado)
        # En un caso real necesitarías mapear las importancias a los nombres de features
        print(f"✅ Modelo con mayor importancia de features: {best_model_name}")
        print(f"   Total de features: {len(importances)}")
        print(f"   Importancia promedio: {np.mean(importances):.4f}")
        print(f"   Desviación estándar: {np.std(importances):.4f}")

def validate_predictions(predictions_df):
    """Valida las predicciones generadas"""
    print("\n🔍 VALIDACIÓN DE PREDICCIONES")
    print("="*40)
    
    # Verificaciones básicas
    total_products = len(predictions_df)
    predictions_positive = (predictions_df['tn'] > 0).sum()
    predictions_zero = (predictions_df['tn'] == 0).sum()
    predictions_negative = (predictions_df['tn'] < 0).sum()
    
    print(f"📊 Total de productos: {total_products}")
    print(f"✅ Predicciones positivas: {predictions_positive} ({predictions_positive/total_products*100:.1f}%)")
    print(f"⚪ Predicciones en cero: {predictions_zero} ({predictions_zero/total_products*100:.1f}%)")
    print(f"❌ Predicciones negativas: {predictions_negative} ({predictions_negative/total_products*100:.1f}%)")
    
    # Verificar valores extremos
    q99 = predictions_df['tn'].quantile(0.99)
    extreme_high = (predictions_df['tn'] > q99 * 2).sum()
    
    print(f"⚠️  Valores extremadamente altos (>2x P99): {extreme_high}")
    
    # Distribución por método
    method_counts = predictions_df['method'].value_counts()
    print("\n📋 Distribución por método:")
    for method, count in method_counts.items():
        print(f"   {method}: {count} ({count/total_products*100:.1f}%)")
    
    return {
        'total_products': total_products,
        'positive_predictions': predictions_positive,
        'zero_predictions': predictions_zero,
        'negative_predictions': predictions_negative,
        'extreme_high': extreme_high,
        'method_distribution': method_counts.to_dict()
    }

def create_submission_summary(predictions_df):
    """Crea un resumen para la submission de Kaggle"""
    print("\n📋 CREANDO RESUMEN DE SUBMISSION")
    print("="*40)
    
    summary_stats = {
        'total_predictions': len(predictions_df),
        'mean_prediction': predictions_df['tn'].mean(),
        'median_prediction': predictions_df['tn'].median(),
        'std_prediction': predictions_df['tn'].std(),
        'min_prediction': predictions_df['tn'].min(),
        'max_prediction': predictions_df['tn'].max(),
        'sum_prediction': predictions_df['tn'].sum(),
        'non_zero_predictions': (predictions_df['tn'] > 0).sum(),
        'zero_predictions': (predictions_df['tn'] == 0).sum(),
    }
    
    print("📊 ESTADÍSTICAS DE SUBMISSION:")
    for key, value in summary_stats.items():
        if isinstance(value, float):
            print(f"   {key}: {value:.2f}")
        else:
            print(f"   {key}: {value}")
    
    # Percentiles
    percentiles = [10, 25, 50, 75, 90, 95, 99]
    print("\n📊 PERCENTILES:")
    for p in percentiles:
        value = predictions_df['tn'].quantile(p/100)
        print(f"   P{p}: {value:.2f}")
    
    return summary_stats

# ============================================================================
# 12. EJECUCIÓN
# ============================================================================

if __name__ == "__main__":
    # Verificar que estamos en el directorio correcto
    if not os.path.exists("../datasets/sell-in.txt"):
        print("❌ Error: No se encuentran los datasets.")
        print("   Asegúrate de estar en el directorio correcto.")
        print("   Los datasets deben estar en ../datasets/")
    else:
        # Ejecutar pipeline completo
        main()
        
        print("\n🎯 SIGUIENTE PASO:")
        print("   Sube el archivo 'kaggle_predictions/kaggle_submission.csv' a Kaggle")
        print("   ¡Buena suerte en la competencia! 🚀")
```
